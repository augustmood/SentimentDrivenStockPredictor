```python
# Calculate daily weighted average
daily_weighted_avg_tsla = tesla_df_cp.groupby('date').apply(
    lambda x: x['weighted_score'].sum() / x['priority'].sum()
).reset_index(name='daily_weighted_avg')

# # Merge the calculated daily weighted averages back into the original DataFrame on 'date'
# tesla_df_cp = pd.merge(tesla_df_cp, daily_weighted_avg_tsla, on='date', how='left')

daily_weighted_avg_tsla['company_name'] = 'Tesla Inc'

daily_weighted_avg_tsla
```

```python
# Calculate daily weighted average
daily_weighted_avg_nvid = nvidia_df_cp.groupby('date').apply(
    lambda x: x['weighted_score'].sum() / x['priority'].sum()
).reset_index(name='daily_weighted_avg')

# # Merge the calculated daily weighted averages back into the original DataFrame on 'date'
# nvidia_df_cp = pd.merge(nvidia_df_cp, daily_weighted_avg_nvid, on='date', how='left')
daily_weighted_avg_nvid['company_name'] = 'NVIDIA Corp'

daily_weighted_avg_nvid
```

```python
# Calculate daily weighted average
daily_weighted_avg_appl = apple_df_cp.groupby('date').apply(
    lambda x: x['weighted_score'].sum() / x['priority'].sum()
).reset_index(name='daily_weighted_avg')

# # Merge the calculated daily weighted averages back into the original DataFrame on 'date'
# apple_df_cp = pd.merge(apple_df_cp, daily_weighted_avg_appl, on='date', how='left')

daily_weighted_avg_appl['company_name'] = 'APPLE Inc'

daily_weighted_avg_appl
```

```python
# Combine the DataFrames
combined_df = pd.concat([daily_weighted_avg_appl, daily_weighted_avg_nvid, daily_weighted_avg_tsla], ignore_index=True)

combined_df['Date'] = pd.to_datetime(combined_df['date'], format='%Y%m%d')

# Drop the original date column if it's no longer needed
combined_df.drop('date', axis=1, inplace=True)

combined_df
```

```python
import pandas as pd

df_stock = pd.read_csv('./stock_data.csv')

df_stock['Date'] = pd.to_datetime(combined_df['Date'], format='%Y%m%d')

df_stock
```

```python
df = pd.merge(df_stock, combined_df, on=['Date', 'company_name'], how='left')

df.to_csv('./data.csv', index=False)
```
