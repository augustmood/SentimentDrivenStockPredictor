```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import yaml
import requests
from datetime import datetime
from matplotlib import rcParams

class StockDataVisualizer:
    def __init__(self, config_path='./config/config.yaml'):
        self.timeout = 15  # Request timeout
        self.config = self.load_config(config_path)

    def load_config(self, config_path):
        with open(config_path, 'r') as file:
            return yaml.safe_load(file)

    def get_stock_data(self, symbol):
        """Fetch stock data for a given symbol."""
        url = f"https://apidojo-yahoo-finance-v1.p.rapidapi.com/market/get-charts"
        params = {
            "region": "US",
            "lang": "en",
            "symbol": symbol,
            "interval": "1d",
            "range": "3mo"
        }
        headers = {
            "X-RapidAPI-Host": self.config['RAPID_API_HOST'],
            "X-RapidAPI-Key": self.config['RAPID_API_KEY'],
            "Content-Type": "application/json"
        }
        response = requests.get(url, headers=headers, params=params, timeout=self.timeout)
        return response.json() if response.status_code == 200 else None

    def extract_timestamps_and_values(self, data):
        """Extract and format timestamps, open, and close values from stock data."""
        timestamps = data["chart"]["result"][0]["timestamp"]
        quotes = data["chart"]["result"][0]["indicators"]["quote"][0]
        open_values = quotes["open"]
        close_values = quotes["close"]

        # Format timestamps
        formatted_dates = [datetime.fromtimestamp(ts).strftime("%m/%d/%Y") for ts in timestamps]
        return formatted_dates, open_values, close_values

    def prepare_data_for_visualization(self, dates, open_values, close_values):
        """Prepare data in a Pandas DataFrame for visualization."""
        events = ['open'] * len(open_values) + ['close'] * len(close_values)
        values = open_values + close_values
        timestamps = dates + dates  # Duplicate for open and close events

        return pd.DataFrame({'Timestamp': timestamps, 'Values': values, 'Events': events})

    def plot_stock_data(self, data_frame, symbol):
        """Plot stock data using seaborn."""
        sns.set_theme(style="darkgrid")
        rcParams['figure.figsize'] = 30, 20
        rcParams['figure.subplot.bottom'] = 0.2

        ax = sns.lineplot(x="Timestamp", y="Values", hue="Events", dashes=False, markers=True, data=data_frame, sort=False)
        ax.set_title(f'Symbol: {symbol}')
        plt.xticks(rotation=45, horizontalalignment='right', fontweight='light', fontsize='xx-small')
        plt.show()

def main():
    visualizer = StockDataVisualizer()
    symbol = input("Enter the stock symbol (at least 2 characters): ").strip()
    
    if len(symbol) > 2:
        stock_data = visualizer.get_stock_data(symbol)
        if stock_data:
            dates, open_values, close_values = visualizer.extract_timestamps_and_values(stock_data)
            data_frame = visualizer.prepare_data_for_visualization(dates, open_values, close_values)
            visualizer.plot_stock_data(data_frame, symbol)
        else:
            print("Failed to fetch stock data.")
    else:
        print("Stock symbol is too short.")

if __name__ == "__main__":
    main()
```
